## Q4. One-hot encoding

```python
january_wo_outliers = january[(january['duration'] >= 1) & (january['duration'] <=60)]
```

```python
january_wo_outliers
```

```python
january_wo_outliers.loc[:,['PULocationID', 'DOLocationID']] = january_wo_outliers.loc[:,['PULocationID', 'DOLocationID']].astype(str)
```

```python
january_wo_outliers.loc[:,['PULocationID', 'DOLocationID']].dtypes
```

```python
jan_dict_list = january_wo_outliers.loc[:,['PULocationID', 'DOLocationID']].to_dict('records')
```

```python
jan_dict_list
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
v = DictVectorizer()
```

```python
X = v.fit_transform(jan_dict_list)
```

```python
X
```

```python
len(v.get_feature_names_out())
```

Q4: 515

## Q5. Training a model

```python
y = january_wo_outliers['duration']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X, y)
```

```python
def compute_rmse(v1, v2):
    return (sum((v1-v2)**2)/len(v1))**0.5
```

```python
y_pred = reg.predict(X)
```

```python
rmse = compute_rmse(y, y_pred)
rmse
```

Q5: 7.65

## Q6. Evaluating the model

```python
def transform_data(df, v):
    df['duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds()/60
    df = df[(df['duration'] >= 1) & (df['duration'] <=60)]
    df.loc[:,['PULocationID', 'DOLocationID']] = df.loc[:,['PULocationID', 'DOLocationID']].astype(str)
    df_dict_list = df.loc[:,['PULocationID', 'DOLocationID']].to_dict('records')
    X = v.transform(df_dict_list)
    y = df['duration']
    return X, y, v
```

```python
X_test, y_test, v_test = transform_data(february, v)
```

```python
len(v_test.get_feature_names_out())
```

```python
y_test_pred = reg.predict(X_test)
```

```python
rmse_test = compute_rmse(y_test, y_test_pred)
rmse_test
```

Q6: 7.81
